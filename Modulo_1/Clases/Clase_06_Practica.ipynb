**Ejercicio 2**

Define una función llamada “harry_potter” que regrese una lista como la siguiente:

`[‘Harry’, ‘Ron’, ‘Hermione’]`

Ayuda: La función recibe los nombres de los personajes

````python
#Pon tu código aquí
def harry_potter(*personaje):
  lista = []
  for i in personaje:
    lista.append(i)
  
  return lista
````

````python
harry_potter('Harry', 'Ron', 'Hermione')
````

La respuesta se ve así 

`['Harry', 'Ron', 'Hermione']`

condiciones importantes
0. Pedir el numero al usuario
1. dividir por todos lo números anteriores
2. mirar(contar) cuantas divisiones son exactas
3. si tiene solo 2 divisiones exactas es primo 

````python
#Paso 0 pedir numero
numero = int(input("Ingresa un numero: "))
s = 0

for i in range(numero):
  num_div = i+1
  if numero % num_div == 0:
    s += 1

if s == 2:
  print("es primo")
else:
  print("no es primo")
````
