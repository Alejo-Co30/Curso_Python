```python
# Usando * (En otros lenguajes, Wildcard o mascarilla ) = Es decir no estoy seguro de la cantidad de argumentos
# que me pase el usuario, se puede recibir valores indiferente de su tipo de dato
# El resultado es un tuppla
def suma(*args):
  s = 0
  for i in args:
#      s = i + s
      s+= i
  return s 
```

```python
suma(5, 7, 1, 2)
```

```python
# Ojo al usar *, la posición es importante - Cuando se uas un argumento sin mascarilla, debemos usarlo al inicio
def comida_favorita(*comida):
  lista_comida = []
  for i in comida:
    lista_comida.append(i)
  return comida
```

```python
# Usando Kwargs (**) argumento llave valor - Iteracion simple del Diccionario
def suma_dic(**kwargs):
  s = 0
  for key in kwargs:
    s += kwargs[key]
  return s
```

```python
suma_dic(a=3, b=10, c=3)
```

```python
# Usando Kwargs (**) argumento llave valor - Iteracion items
def suma_dic(**kwargs):
  s = 0
  for key, value in kwargs.items():
    print (key, '=', value)
    s += value
  return s
```

```python
suma_dic(a=3, b=10, c=3)
```

```python
# Añadir comentarios es importante, para documentar la funcionalidad de alguna función, esto se usa con triple comilla
# Estas deben estar dentro de la función
# Y usando la función help para mostrar los coemntarios de la función - Ojo si usas # te tare igual el mensaje
#Este tipo de comentarios se conoce como docstring

def suma_dic2(**kwargs):
  """ Hola es un mensaje
  """
  s = 0
  for key, value in kwargs.items():
    print (key, '=', value)
```

```python
help(suma_dic2)
```
