Funciones: 
* Permite disminuir las repeticiones desde el llamado de la misma.
* Son bloques de código, que luego seran invocados.
* Si inicia con def ´"Palabra reservada"
* Por buenas prácticas, el nombre de las mismas debe ser claro "Descriptivo", en base a su función. 
* Luego del nombre se acompaña con un parentésis y las mismas pueden recibir un argumento.
* Continuamos con dos puntos, y luego de la identación viene la sentencia.
* Por último se usa la palabra reservada el Return.

Caracteristicas:
- Reusabilidad, modulavilidad 

```python
# Función sin argumento
# Aquí no uso el print(), ya que el return no permite devolver un método
def mensaje_bienvenida():
  return "Hola Mundo"

mensaje_bienvenida()
```

```python
# Función con argumento
def mensaje_bienvenida_arg(nombre):
  return "Hola Mundo" + nombre

mensaje_bienvenida_arg("andres")
```

Crear una función que diga, cual es tu comida favorita

```python
# Con un sólo argumento
def comida_favorita(comida):
  return "Tu Comida favorita es:" + comida

comida_favorita("Arroz")
```

```python
# Con dos argumentos
def comida_favorita2(comida, nombre):
  return "Hola " + nombre + " Tu Comida favorita es: " + comida

comida_favorita2("Arroz", "María")
```

```python
#Usando input()
comida = input()
def comida_favorita2(comida, nombre):
  return "Hola " + nombre + " Tu Comida favorita es: " + comida

comida_favorita2(comida, "María")
```
