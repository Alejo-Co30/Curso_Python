```python
# *args -> tupla
# **kwargs (keyword arguments) -> diccionario
```

```python
def suma(divisor, *nums):
  # nums será una tupla
  print(nums)
  for numero in nums:
    print(numero/divisor)
  
suma(3,4,5,6)
```

```python
def imprimir_datos(**datos):
  # datos será un diccionario
  print(f"Hola {datos['nombre']} tu edad es {datos['edad']}")
```

```python
imprimir_datos(nombre="sebastian", edad=21, ciudad="tunja")
```

```python
def usando_args(num1, num2, num3, otro_num):

  print(num1, num2, num3, otro_num)


usando_args(2, *[3,4,5])
```

```python
def usando_kwargs(nombre, apellido):

  print(nombre, apellido)
usando_kwargs(**{'nombre': 'miguel', 'apellido': 'toro'})
```

```python
def combinar(x, *args, **kwargs):
  print(x)
  print(args)
  print(kwargs)
combinar(3,5,6,nombre="pepe",peso=70, otro_nombre="juan")
```

```python
def combinar(x, *args):
  print(x)
  print(args)
  
combinar(3,5,6)
```

```python
def combinar(x, **kwargs):
  print(x)
  print(kwargs)
  print(kwargs['nombre'])
combinar(3, nombre="pepe", peso=70, otro_nombre="juan")
```

```python
def imprimir_datos(**datos) -> int:
  """ imprime datos
    Args:
      datos(str): esto es un diccionario de datos personales

    Return:
      int: 
  """
  for llave, valor in datos.items():
    print(llave, valor)
  return 123
```

```python
imprimir_datos(nombre="karol", ciudad="Bogota")
```
