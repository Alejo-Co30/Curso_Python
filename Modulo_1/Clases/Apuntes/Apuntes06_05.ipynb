Para Pytho, un conjunto en un Objeto de tipo de dato Conjunto
- Los elementos de los conjuntos son desordenados
- Los elementos almacenados son inmutable - No pude cambiar sus elementos, pero si pueden agregar o eliminar elementos
- Los elementos son unicos, sin duplicados. - True y 1 se condiferan el mismo elemento por ende son duplicados

```python
#Primera forma - Usando el método set
#Este método transforma cualquier tipo iterable en SET


print(type([1, 2, 3, 3, 4]))
print(type(set([1, 2, 3, 3, 4])))

lis = set([1, 2, 3, 3, 4])
print(lis)

# Segunda forma - usando llaves
lis = {4,9,7,4}
print(lis)
```

```python
# Recorriendo los SET

lis = {1,4,7,9}

for e in lis:
  print(e)

# Determinar el tamaño de un set o conjunto

print(len(lis))

# Determinando si un elemento está en un conjunto

print(1 in lis)

if 0 in lis:
  print("Es true")
else:
  print("Es false")
```

```python
# Operaciones de los conjuntos  - Método
# Unión: Une los dos conjuntos, sacando los repetidos

x1 = {'foo', 'bar', 'ba'}
x2 = {'fo1', 'ba1', 'ba2'}

x1.union(x2)
 # Usando "|" x1 | x2
```

```python
x3 = {'foo', 'bar', 'ba1'}
x4 = {'fo1', 'ba1', 'ba2'}

# intersección: intersection - Elemento que se repite para ambos conjuntos
# ba1 se encuntra en mavos conjuntos

x3.intersection(x4)
 # Usando "&" 
x3 & x4
```
