```python
def suma(*args):
  s = 0

  for i in args:
    s += i #s = i + s
  
  return s
```

```python
def suma_dict(**kwargs):
  s = 0
  for key in kwargs:
    s += kwargs[key]
  return s
```

```python
suma_dict(a=3,b=10,c=3)
```
