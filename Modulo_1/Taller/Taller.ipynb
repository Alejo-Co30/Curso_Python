```python
# Crea un pequeño script que sea capaz de calcular cuantos segundos hay en una hora y guárdalo en una variable.
Hora = int(input("Ingrese una hora: "))
ResulHora = Hora*3650
print("En",Hora,"horas hay",ResulHora)
```

```python
# Usando la variable de la pregunta anterior, calcula ¿cuántos segundos hay en un día?
# Comnetario: Al calcularse el ejercicio anterio por 3650 segundos en horas, 

if (Hora >= 24):
  ResulDia = (Hora * 1)/24
  Resultado = ResulDia*3650
  print("En", ResulDia,"dias hay",Resultado, "segundos")
else: 
  print(Hora,"es menor a un día")
```

```python
from IPython.utils.path import random
# Crea un script en donde le asignes un valor aleatorio a una variable y que esta sea
# usada para comparar un número solicitado al usuario. Escriba las pruebas
# condicionales correspondientes

variable = random.randint(3, 9)
print("Variable del sistema",variable)

varUser = input("Ingrese un valor ")
# Condicional, es un valor par o impar - If
print("Usando If")
resOne = int(varUser)%variable

if (resOne == 0) :
  print ("El valor ingresado", varUser, "es par de",variable)
else:
  print ("El valor ingresado", varUser, "es impar de",variable)

# Condicional, siempre que la variable del sistema  se superior o igual a la variable del usuario - While
print("Usando While")
while variable >= 0 and int(varUser) >= 0:
  if (variable >= int(varUser)):
    print("La variable", variable,"del sistema es mayor a",varUser)
    break
  else:
    print("La variable", variable,"del sistema es menor a",varUser)
    break
```

```python
# Determine cuál sería el peso de una persona si estuviera en cualquiera de los
# planetas del sistema solar. El usuario debe tener la posibilidad de escoger el planeta
# que desea evaluar

thePlanets = {
    1: "Tierra",
    2: "Marte",
    3: "Jupiter",
    4: "Pluton"
}
theWeith = {
    1: 40,
    2: 13,
    3: 24,
    4: 34
}

print("Los planetas son\n 1. ", thePlanets[1],"\n 2. ", thePlanets[2],"\n 3. ", thePlanets[3],"\n 4. ", thePlanets[4])
X = int(input("\nElige un planeta ingresando el número del planeta: "))
Y = int(input("\nSeñor usuario ingrese su peso actual: "))

for i in theWeith:
  if X == i:
    Z = Y * theWeith[X]
    print('Su peso es: ', Z, 'Kg', )
```
