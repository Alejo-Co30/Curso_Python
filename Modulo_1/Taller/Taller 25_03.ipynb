**Ejercicio 2**

Define una función llamada “harry_potter” que regrese una lista como la siguiente:

`[‘Harry’, ‘Ron’, ‘Hermione’]`

Ayuda: La función recibe los nombres de los personajes

````python
#Pon tu código aquí

def harry_potter(nom_1, nom_2, nom_3):
  lista = []
  lista.append(nom_1)
  lista.append(nom_2)
  lista.append(nom_3)
  return lista
````

````python
harry_potter("Harry","Ron", "Hermione")
````

````python
#Pon tu código aquí

def harry_potter(nom_1, nom_2, nom_3):
  lista = []
  lista.extend([nom_1, nom_2, nom_3])
  
  return lista
````
