```python
# A y B es una referencia
def operacion(a, b):
  multi = a * b
  resta = a - b
  return multi, resta
num_1 = int(input("Ingrese el primer valor "))
num_2 = int(input("Ingrese el segundo valor "))
operacion(num_1,num_2)
```
