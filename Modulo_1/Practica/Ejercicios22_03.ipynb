Accediendo a los datos

```python

dic = {
    "Nombre": "María",
    "Edad": 23,
    "opciones": [1,2,3]
}

# Accediendo a los valores desde su clave

x = dic["Nombre"]
print(x)

# Accediendo a una lista del Dic

y = dic["opciones"]
print(y)
```

```python
dic = {
    "Nombre": "María",
    "Edad": 23
}

# Accediendo con un método
z = dic.get("Nombre")
print(z)
```

```python
dic = {
    "Nombre": "María",
    "Edad": 23
}
# Accediendo a las claves - Este método es una vista de las claves, por lo tanto se actualiza
# segun cambios en el diccionario, devuelve una lista

v =  dic.keys()
print(v)
```

```python
dic = {
    "Nombre": "María",
    "Edad": 23
}
# Accediendo a los valores - Este método es una vista de los valores, por lo tanto se actualiza
# segun cambios en el diccionario, devuelve una lista

w = dic.values()
print(w)
```

```python
dic = {
    "Nombre": "María",
    "Edad": 23
}

# Accediendo a los artículos- Este método es una vista de los elementos, por lo tanto se actualiza
# segun cambios en el diccionario, devuelve una lista de tuplas

u = dic.items()
print(u)
```

```python
dic = {
    "Nombre": "María",
    "Edad": 23
}

# Cambiando los valores si no existe la clave valor, igual se inyecta
# en el diccionario

print(dic["Nombre"])
dic["Nombre"] = "Pedro"
print(dic["Nombre"])

dic["Apellido"] = "Pedro"
print(dic)


```

```python
dic = {
    "Nombre": "María",
    "Edad": 23
}

# Cambiando valores usando un método - si no existe la clave valor, igual se inyecta
# en el diccionario

print(dic["Nombre"])
dic.update({"Apellido":"Pedro"})
print(dic["Nombre"])
print(dic)
```

```python
dic = {
    "Nombre": "María",
    "Edad": 23
}

# Eliminando elementos - pop elimina desde la clave

print(dic)
dic.pop("Nombre")
print(dic)

dic["Nombre"] = "Maria"
print(dic)

# Eliminando usando popitem - elimina el último elemento insertado
dic.popitem()
print(dic)

# Elimando usando la palabra reservada del

dic.update({"Nombre":"Carlos"})
print(dic)
del dic["Edad"]
print(dic)
```

```python
dic = {
    "Nombre": "María",
    "Edad": 23
}

# Método para vaciar el diccionario

print(dic)
dic.clear()
print(dic)
```

```python
dic = {
    "Nombre": "María",
    "Edad": 23
}

# Copiando diccionarios con el método copy()

print(dic)
dicCopy = dic.copy()
print(dic)
print(dicCopy)

# Copiando diccionarios con la función dict
dicCopyCopy = dict(dicCopy)
print(dicCopyCopy)
```

```python
dic = {
    "Nombre": "María",
    "Edad": 23
}

# Métodos especiales setDefault() - Argumento es Clave y valor 
x = dic.setdefault("Nombre", "Maria")
print(x)

#- El valor el opcional
y = dic.setdefault("Nombre")
print(y)

# Si no existe, lo inyecta en el Diccionario
print(dic)
dic.setdefault("Apellido", "Carvajal")
print(dic)

# Métodos especiales fromkeys - Crea un diccionario apartir de un valor y claves
# especificas

a = ('Nombre', 'edad')
b = (0)

OneDic = dict.fromkeys(a,b)
print(OneDic)
```

```python
dic = {
    "Nombre": "María",
    "Edad": 23
}

# Usando Bucles - Claves

for x in dic:
  print(x)

# Usando Bucles - Valores

for x in dic:
  print(dic[x])

# Usando un método - Values

for x in dic.values():
  print(x)

# Usando un método - keys

for x in dic.keys():
  print(x)

# Usando un método - items

for x, y in dic.items():
  print(x, y)
```

```python
# Diccionarios anidados
dic = {
    "Nombre": "María",
    "Edad": 23
}

dic2 = {
    "Apellido": "Carvajal",
    "Altura": 1.50
}

person = {
    "dic":dic,
    "dic2":dic2
}

print(person["dic2"]["Altura"])
```
